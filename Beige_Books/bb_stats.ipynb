```python
import os as os
```

```python
directory_path = "C:/Users/MCOB PHD 14/Dropbox/Charlie's Dissertation/Beige Books"
```

```python
text_data = []
```

```python
# Loop through the files in the directory
for filename in os.listdir(directory_path):
    if filename.endswith(".txt"):
        with open(os.path.join(directory_path, filename), 'r', encoding='utf-8') as file:
            text_data.append(file.read())
```

```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter

nltk.download('punkt')

# Initialize an empty list to store word counts for each document
word_counts = []

```

```python
# Tokenize and count words in each document
for document in text_data:
    tokens = word_tokenize(document)
    word_counts.append(len(tokens))
```

```python
import numpy as np

# Calculate summary statistics
mean_word_count = np.mean(word_counts)
median_word_count = np.median(word_counts)
max_word_count = np.max(word_counts)
min_word_count = np.min(word_counts)

print("Mean Word Count:", mean_word_count)
print("Median Word Count:", median_word_count)
print("Max Word Count:", max_word_count)
print("Min Word Count:", min_word_count)
```

```python
import pandas as pd
wordsDF = pd.DataFrame(word_counts, columns = ['Word_Count'])
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# Sample text data
text = text_data[0]

# Tokenize the text into words
words = nltk.word_tokenize(text)

# Remove stopwords
filtered_words = [word for word in words if word.lower() not in stopwords.words('english')]

# Join the filtered words back into a sentence (if needed)
filtered_text = ' '.join(filtered_words)
filtered_text_token = nltk.word_tokenize(filtered_text)
```

```python
filtered_text_token
```

```python
# Count the occurrences of each word
word_counter = Counter(filtered_text_token)

# Get the most common words and their counts (e.g., top 10)
most_common_words = word_counter.most_common(25)  # Change 10 to the desired number of top words

# Print the most common words and their counts
for word, count in most_common_words:
    print(f"{word}: {count}")
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon (if you haven't already)
nltk.download('vader_lexicon')
```

```python
# Initialize the sentiment analyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Your document, 4,400 is from Atlanta in Dec. 3, 2008 and was extremely negative sentiment
document = text_data[4400]

# Perform sentiment analysis
sentiment_scores = analyzer.polarity_scores(document)

# Interpret the sentiment scores
compound_score = sentiment_scores['compound']

if compound_score >= 0.05:
    sentiment = 'Positive'
elif compound_score <= -0.05:
    sentiment = 'Negative'
else:
    sentiment = 'Neutral'

print("Sentiment:", sentiment)
print("Compound Score:", compound_score)
```
